## Comparing Algorithms

Consider two algorithms, one with runtime $T_a(n) = 5n + 1000$, and the other with runtime $T_b(n) = 10n+2$. Which is better?

```python
n = np.arange(0,1000)
a = 5*n + 1000
b = 10*n + 2

fig = plt.figure(figsize=(6,4))
plt.plot(n,a,label='$T_a$')
plt.plot(n,b,label='$T_b$')
plt.legend(fontsize=16)
```

What if the two algorithms are $T_a(n) = 1000n^2$ and $T_b(n) = n^8$?

```python
n = np.arange(0,4,0.01)
a = 1000*np.square(n)
b = n**8

fig = plt.figure(figsize=(6,4))
plt.plot(n,a,label='$T_a$')
plt.plot(n,b,label='$T_b$')
plt.legend(fontsize=16)
```

## How do we quantify this?

We care about the number of elementary operations on worst-case inputs as a function of input size n, when n tends to infinity. In other words, we care about how quickly the runtime grows as n gets bigger.

### Suppose we have a function with T(n) = 2n + 10. Can we prove that this is O(n)?

Suppose we set c=3. Can we find an n_0 that satisfies the condition? We must solve:

            2n + 10 <= 3n
            -n + 10 <= 0
            n >= 10
            
So, n_0 >= 10 satisfies the condition, meaning that T(n) is O(n).

### Generally speaking, we can eyeball Big-O:

If T(n) is a polynomial of degree d, such that $T(n)$ = $a*n^d + b*n^{d-1}$ + ... + $w*n + z$, then by definition $T(n)$ is $O(n^d)$
