# For Loops

### Example 1:

Write code for a function, printAllElements(array), which given an array prints all elements followed by an exclamation point.

```python
def printAllElements(array):
    for x in array:
        print(f"{x}!")
        
printAllElements(["hi","hello","howdy"])
```

### Example 2:

Write code for a function, printSquareRoot(N), which given a number N, prints the square root of all numbers between 1 and N.

```python
import numpy as np

def printSquareRoot(N):
    for x in range(1,N):
        print(np.sqrt(x))

printSquareRoot(4)
```

### Example 3:

Write code for a function, printIndex(array, value), which given an array and a value returns the index of the array that contains the specified value. If the value does not exist in the array, print 'Value not found.' and return None.

```python
def printIndex(array, value):
    for i,x in enumerate(array):
        if x==value:
            return i
    print("Value not found.")

printIndex(["a","c","r"],"d")
```

### Example 4-a:

Write code for a function, findConsecutiveMax(array), which given
an array of characters, returns the character that occurs the greatest number of
consecutive times. If multiple characters appear consecutively the maximum number of times, your function should return the character whose sequence comes first in the array.

```python
def findConsecutiveMax(array):
    maxCharacter=""
    numMax=0
    lastChar=""
    numCurrent=1
    for i,x in enumerate(array):
        if x==lastChar:
            numCurrent=numCurrent+1
        else:
            numCurrent=1
        if numCurrent>numMax:
            numMax=numCurrent
            maxCharacter=x
        lastChar=x
    return maxCharacter
    
findConsecutiveMax(["z","a","a","z","y","y","y","z","z","z"])
```

### Example 1:

Write a function, computeFactorial(N), which given a number N uses a while loop to compute and return N!.

```python
def computeFactorial(N):
    i = 0
    output = 1
    while i < N:
        output = output*(N-i)
        i = i + 1
    return output

computeFactorial(0)
```

### Example 2-a:

Write a function, getNumDivisions(N), which given a number N uses a while Loop to compute and return how many times N can be divided evenly by 2.

```python
def getNumDivisions(N):
    numTimes = 0
    while N%2 == 0:
        N = N/2
        numTimes += 1
    return numTimes

getNumDivisions(16)
```

### Example 2-b:

Adjust getNumDivisions(N) to print out the first non-integer number produced by continuously dividing N by 2. The print statement should read 'Dividing N in half {x} times results in {y}'

```python
def getNumDivisions(N):
    numTimes = 0
    while N%2 == 0:
        N = N/2
        numTimes += 1
    N = N/2
    numTimes += 1
    print(f"Dividing N in half {numTimes} times results in {N}.")

getNumDivisions(18)
```
