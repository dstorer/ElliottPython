# What is a tree?

A tree is a code implementation of hierarchy, consisting of nodes with parent/child relationships. The file system on your computer is a great example of a tree.

Some useful tree terminology:

- <b>Root</b>: node without a parent (A)
- <b>Internal node</b>: node with at least one child (A, B, C, F)
- <b>Leaf (external node)</b>: node without children (E, I, J, K, G, H, D)
- <b>Parent node</b>: node immediately above a given node (parent of C is A)
- <b>Child node</b>: node(s) immediately below a given node (children of C are G and H)
- <b>Ancestors of a node</b>: parent, grandparent, grand-grandparent, etc. (ancestors of G are C, A)
- <b>Descendant of a node</b>: child, grandchild, grand-grandchild, etc.
- <b>Depth of a node</b>: number of ancestors (I has depth 3)
- <b>Height of a tree</b>: maximum depth of any node (tree with just a root has height 0, this tree has height 3)
- <b>Subtree</b>: tree consisting of a node and its descendants

## Associated Methods

Tree methods
- int <b>size( )</b>: returns the number of nodes
- boolean <b>isEmpty( )</b>: returns true if the tree is empty
- Node <b>root( )</b>: returns the root of the tree
    
Node methods:
- Node <b>parent( )</b>: returns the parent of the node
- Node[ ] <b>children( )</b>: returns the children of the node
- boolean <b>isInternal( )</b>: returns true if the node has children
- boolean <b>isExternal( )</b>: returns true if the node is a leaf
- boolean <b>isRoot( )</b>: returns true if the node is the root

## Binary Trees

Nodes have at most 2 children, left and right. 

Binary trees can be defined recursively, where each tree is just a root node with at most 2 children, and each child is itself a binary tree.

In addition to the above methods, a binary tree should also have:
- Node <b>left( )</b>: returns the left child if it exists, else None
- Node <b>right( )</b>: returns the right child if it exists, else None
- Node <b>hasLeft( )</b>: returns TRUE if node has left child
- Node <b>hasRight( )</b>: returns TRUE if node has right child
- Node <b>setLeft( )</b>: creates and returns a left child
- Node <b>setRight( )</b>: creates and returns a right child
