```python
from matplotlib import pyplot as plt
import matplotlib
import numpy as np
import random
```

```python
RAs = np.linspace(-180,180,100)
Decs = np.linspace(-90,90,100)
Fints = np.linspace(0,100,100)
```

```python
random.shuffle(RAs)
random.shuffle(Decs)
random.shuffle(Fints)
```

```python
def sourceMap(RAs, Decs, Fints, unit='deg'):
    ## select/unit stuff
    fig = plt.figure(figsize=(12,8))
    cmap = matplotlib.cm.get_cmap('viridis')
#     Fints = np.log10(Fints)
#     sourcecolor = cmap(Fints/np.max(Fints))
    im = plt.scatter(RAs,Decs,c=Fints, norm=matplotlib.colors.LogNorm())
    x_ticks = [int(i) for i in np.linspace(-200,200,11)]
    y_ticks = [int(i) for i in np.linspace(-110,110,11)]
    plt.xticks(x_ticks)
    plt.yticks(y_ticks)
    plt.xlabel(f'Right Ascension ({unit})')
    plt.ylabel(f'Declination ({unit})')
    plt.title(f'Source Map at ()')
    plt.colorbar(im)
```

```python
sourceMap(RAs,Decs,Fints)
```
